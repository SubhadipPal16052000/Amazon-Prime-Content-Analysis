#### MISSING VALUES

```python
# Checking for null value.
#df.isnull().sum()
df.isna().sum()
```

```python
# Count of unique values.
df.nunique()
```

```python
#Drop null value from Description column.
df.dropna (subset=['description'], inplace = True)
```

```python
df.isnull().sum()
```

```python
#Transefrom the mode of age certification in zero where NAN value is present.

df['age_certification'].mode()[0] # mode of age certification

df['age_certification'].fillna(df['age_certification'].mode()[0], inplace = True)#Replacing all null values of age_certification with mode
```

```python
df.isnull().sum()
```

### The 'seasons' column has null values because its only applicapable to tv show, movies don't have any seasons so these entries would naturally be null. Instead of replacing this  row , replace them with zero.  

```python
# Replacing null values from seasons
df['seasons'].fillna(0, inplace = True)
```

```python
df.isnull().sum()
```

Here 'imdb_id' is unique for Movies and tv show, So we will drop all null values in 'imdb_id' columns.

```python
# Drop all null values from imdb_id.
df.dropna(subset =['imdb_id'], inplace = True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

'imdb_score' represents the  rating of a movie or tv show, removing the  null values in 'imdb-score' might lead to the data loss. Its better to replace them with mean.



```python
# Removing the null values in imdb_score might lead to data loss, its better to replace to them with mean.
round(df['imdb_score'].mean(),1) # Generalizing the mean value
#Replaceing the  null values in imdb_score with mean of imdb_score.
df['imdb_score'].fillna(round(df['imdb_score'].mean(),1), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

'imdb_votes' represent the number of votes a movie or tv show has received. Missing  values in the 'imdb_votes' indicate movie is  relatively new or has not gained much more attention. Its better replace null value with zero.

```python
# Missing value on imdb_votes may indicate movie ie raltively new or has not gain attention much more,
# So better approche to replace null value with zero.
df['imdb_votes'].fillna(0, inplace = True)
```

```python
df.isnull().sum()
```

```python
round(df['tmdb_popularity'].mean(),2) # Generalizing the mean value
#Replaceing the  null values in tmdb_populartity with mean of tmdb_popularity
df['tmdb_popularity'].fillna(round(df['tmdb_popularity'].mean(),2), inplace=True)
```

```python
df.isnull().sum()
```

```python
round(df['tmdb_score'].mean(),1)
```

```python
#Replaceing the  null values in tmdb_score with mean of tmdb_score
df['tmdb_score'].fillna(round(df['tmdb_score'].mean(),1), inplace=True)
```

```python
df.isnull().sum()
```

Missing value in the 'character' column indicate that the information about the character played by a persons in the movie or tv show is not available in the data set. So replace them by 'Unknown'.   

```python
# Missing values in character column replace by unknown.
df['character'].fillna('unknown', inplace= True)
```
